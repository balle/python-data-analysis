```python
data = pd.read_csv("data/countries.csv")
usa = data[data.country == 'United States']
china = data[data.country == 'China']

plt.plot(usa.year, usa.population / 10 ** 6)
plt.plot(china.year, china.population / 10 ** 6)
plt.legend(["USA", "China"])
plt.show()
```
